```python
import sys

PRICE_UNIT = 100

class message:
    title = '##### %s 자판기 입니다 #####'
    product = f'%s:%s(%s원)'
    insert_coin = '금액을 넣어주세요: '
    short_coin = f'금액이 부족합니다.\n잔액은 %s원 입니다.'
    select_product = '상품 번호를 선택해 주세요: '
    select_fault = '잘못된 숫자입니다.'
    product_out = f'선택하신 상품은 %s 입니다.\n잔액은 %s원 입니다.\n감사합니다.'

class Product:
    product_type ={'1':'카푸치노','2':'카페라떼','3':'아메리카노'}
    product_val = {'1':300,'2':400,'3':200}

class CoffeeVM(Product):
    name = '커피'
    
    def __init__(self):
        print(message.title %self.name)
    
    def run(self):
        while True:
            try:
                input_coin = float(input(message.insert_coin))
            except ValueError:
                print(message.select_fault)
            else:
                self.select(input_coin)
           
    def select(self,coin):
        description =[]
        for selection, item in Product.product_type.items():
            price = self.get_product_value(selection)
            description.append(f'{selection}:{item}({str(price)}원)')
        print(description)
        input_product = input(message.select_product)
        price = self.get_product_value(input_product)
        
        if price:
            p_name = self.get_product_name(input_product)
            self.payment(coin,p_name,price)
        else:
            print(message.select_fault)
            self.select(coin)
    
    def get_product_value(self,product):
        return_val = 0
        for selection, value in Product.product_val.items():
            if selection == product:
                return_val = value
        return return_val
    
    def get_product_name(self,product):
        for selection, name in Product.product_type.items():
            if selection == product:
                return name
    
    def payment(self,coin,name,value):
        coin_val = coin/100 * PRICE_UNIT
        if coin_val >= value:
            balance = coin_val - value
            print(message.product_out %(name,int(balance)))
        else:
            print(message.short_coin %int(coin_val))
        self.run()
        

class SnackVM(CoffeeVM):
    name = '과자'
    
    def __init__(self):
        Product.product_type = {'1':'감자칩','2':'젤리','3':'초콜렛','4':'나초'}
        Product.product_val = {'1':1500, '2':700, '3':1000, '4':1200}
        print(message.title %self.name)
```

```python
if __name__ == '__main__':
    print('1:커피','2:과자')
    select_vm = input("자판기를 선택해 주세요:").strip()
    
    if select_vm == '1':
        vm = CoffeeVM()
        
    elif select_vm == '2':
        vm = SnackVM()
        
    else:
        print('잘못된 번호입니다. 다시 선택해 주세요.')
        sys.exit(-1)
        
    try:
        vm.run()
    except KeyboardInterrupt as e:
        print('판매 종료.')
```
