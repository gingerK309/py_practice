```python
import math
from math import pow, ceil
from matplotlib import pyplot as plt
```

```python
GRAVITY = 9.81
INTERVAL = 0.001
```

```python
def time_arr(start,end,interval):
    times=[]
    while start<end:
        times.append(start)
        start += interval
    return times

def draw_parabola(x_velocity,y_velocity):
    time = 2 * y_velocity/GRAVITY
    distance = x_velocity * time
    intervals = time_arr(0,time,INTERVAL)
    
    x=[]
    y=[]
    
    for t in intervals:
        x.append(x_velocity*t)
        y.append(y_velocity*t - 0.5*GRAVITY*(t**2))
    plt.plot(x,y)
    plt.xlabel(f'{str(x_velocity)}m/s')
    plt.ylabel(f'{str(y_velocity)}m/s')
    plt.title(f'flying time is {str(math.ceil(time))}s, distance is {str(math.ceil(distance))}m')
```

```python
if __name__ == '__main__':
    try:
        x_velocity = float(input('x 벡터 속도를 입력해주세요(m/s): '))
        y_velocity = float(input('y 벡터 속도를 입력해주세요(m/s): '))
    except ValueError:
        print('올바른 값을 입력해 주세요.')
    else:
        draw_parabola(x_velocity,y_velocity)
        plt.show()
```
