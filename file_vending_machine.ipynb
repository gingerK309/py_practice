```python
import sys

PRICE_UNIT = 100

class message:
    title = '##### %s 자판기 입니다 #####'
    product = '%s:%s(%s원)'
    insert_coin = '금액을 넣어주세요: '
    short_coin = '금액이 부족합니다.\n잔액은 %s원 입니다.'
    select_product = '상품 번호를 선택해 주세요: '
    select_fault = '잘못된 숫자입니다.'
    product_out = '선택하신 상품은%s 입니다.\n잔액은 %s원 입니다.\n감사합니다.'

class Product:
    product_type = []
    product_val = []

class CoffeeVM(Product):
    info_file = 'coffee.txt'
    name = '커피'
    
    def __init__(self):
        print(message.title %self.name)
    
    def run(self):
        self.set_products()
        while True:
            try:
                input_coin = float(input(message.insert_coin))
            except ValueError:
                print(message.select_fault)
            else:
                self.select(input_coin)
    
    def set_products(self):
        Product.product_type = []
        Product.product_val = []
        with open(self.info_file,'r',encoding='UTF-8') as file:
            for line in file:
                list = line.strip('\n').split(',')
                Product.product_type.append((list[0]+','+list[1]))
                Product.product_val.append((list[0]+','+list[2]))

    def select(self,coin):
        description = []
        for line in Product.product_type:
            list = line.split(',')
            price = self.get_product_value(list[0])
            description.append(f'{list[0]}:{list[1]}({str(price)}원)')
        print(description)
        input_product = input(message.select_product)
        price = self.get_product_value(input_product)
        
        if price:
            p_name = self.get_product_name(input_product)
            self.payment(coin,p_name,price)
        else:
            print(message.select_fault)
            self.select(coin)
    
    def get_product_value(self,product):
        return_val = 0
        for line in Product.product_val:
            list = line.split(',')
            if list[0] == product:
                return_val = list[1]
        return int(return_val)
    
    def get_product_name(self,product):
        for line in Product.product_type:
            list = line.split(',')
            if list[0] == product:
                return list[1]
    
    def payment(self,coin,name,value):
        coin_val = coin/100 * PRICE_UNIT
        if coin_val >= value:
            balance = coin_val - value
            print(message.product_out %(name,int(balance)))
        else:
            print(message.short_coin %int(coin_val))
        self.run()
        

class SnackVM(CoffeeVM):
    info_file = 'snack.txt'
    name = '과자'
```

```python
if __name__ == '__main__':
    print('1:커피','2:과자')
    select_vm = input("자판기를 선택해 주세요:").strip()
    
    if select_vm == '1':
        vm = CoffeeVM()
        
    elif select_vm == '2':
        vm = SnackVM()
        
    else:
        print('잘못된 번호입니다. 다시 선택해 주세요.')
        sys.exit(-1)
        
    try:
        vm.run()
    except KeyboardInterrupt as e:
        print('판매 종료.')
```
